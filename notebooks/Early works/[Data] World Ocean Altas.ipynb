```python
import glob
import numpy as np
import netCDF4 as nc
import matplotlib.pyplot as plt
from __future__ import division
from __future__ import print_function
from mpl_toolkits.basemap import cm
from mpl_toolkits.basemap import Basemap
from mpl_toolkits.basemap import shiftgrid
% matplotlib inline
```

```python
def draw_ORCA2_Atlantic(lon, lat, var, clev, var_name, CMap):
    '''
        Draw NEMO forcing data on a global scale
    '''
    #fig    
    fig=plt.figure(figsize=(14, 14))
    axes=plt.gca()
    #left
    proj=Basemap(projection='mill', llcrnrlat=60, urcrnrlat=85,\
            llcrnrlon=-30, urcrnrlon=10, resolution='l')
    parallels=np.arange(0, 90, 10); meridians=np.arange(-180, 180, 10)
    proj.drawparallels(parallels, labels=[1, 0, 0, 0], fontsize=10)
    proj.drawmeridians(meridians, labels=[0, 0, 0, 1], fontsize=10)
    proj.drawcoastlines(linewidth=1.5, linestyle='-', color='k', zorder=2)
    proj.drawlsmask(land_color=[0.5, 0.5, 0.5], ocean_color='None', lsmask=None, zorder=1, resolution='c')
    x, y=proj(lon, lat)
    CS=proj.contourf(x, y, var, clev, cmap=CMap, extend='both', zorder=2)
    proj.contour(x, y, var, clev, colors='k', linewidths=2.0, zorder=2)
    # colorbar
    CBar=proj.colorbar(CS, location='right', size='5%', pad='10%')
    CBar.set_label(var_name, fontsize=10, fontweight='bold')
    CBar.ax.tick_params(axis='y', length=17.5)
    return fig, axes, proj
```

```python
JJA_name=glob.glob('_data/WOA/*s13*.nc')
print("found file: \n{}".format(JJA_name))
DJF_name=glob.glob('_data/WOA/*s15*.nc')
print("found file: \n{}".format(DJF_name))
```

```python
JJA_obj=nc.Dataset(JJA_name[0])
DJF_obj=nc.Dataset(DJF_name[0])
print('Variables in {}:\n==========\n{}'.format(JJA_name[0], JJA_obj.variables.keys()))
```

```python
lon=JJA_obj.variables['lon'][:]
lat=JJA_obj.variables['lat'][:]
JJAsss=JJA_obj.variables['s_an'][0, 0, :, :]
DJFsss=DJF_obj.variables['s_an'][0, 0, :, :]
longitude, latitude = np.meshgrid(lon, lat)
```

```python
clev=np.arange(29, 38, 1)
#CMap=cm.GMT_no_green
CMap=plt.cm.RdBu_r
CMap.set_over(CMap(np.arange(256))[-1, 0:3])
CMap.set_under(CMap(np.arange(256))[0, 0:3])
var_name='Salinity, PSU'
fig, ax, proj = draw_ORCA2_Atlantic(longitude, latitude, DJFsss, clev, var_name, CMap)
title='DJF Sea Surface Salinity (1996-2005)'
ax.set_title(title, fontsize=12, fontweight='bold')
fig.set_size_inches(7, 14)
plt.savefig('FIG5.eps', dpi=600, orientation='portrait', papertype='a4', format='eps',
            transparent=False, bbox_inches='tight', pad_inches=0)
fig.canvas.draw()
```
