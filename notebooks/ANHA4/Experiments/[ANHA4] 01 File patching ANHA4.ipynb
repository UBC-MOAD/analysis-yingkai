# File patching ANHA4

This notebook discribes the Pythonic way to resolve the discrepancies between Xianmin's ANHA4-EXH001 forcing and Susan's `fldread.F90`.

Operating sequence:

* Link the original files
* Create rename.py, fix_2004.py, fix_2008.py, fix_2012.py
* python rename.py > rename.sh
* python fix_2004.py > fix_2004.sh
* python fix_2008.py > fix_2008.sh

# Rename

the end of every fdays --> the beginning of every five days

```python
names = ['ANHA4-EXH005_gridT_', 'ANHA4-EXH005_gridU_', 'ANHA4-EXH005_gridV_',
'ANHA4-EXH005_gridW_', 'ANHA4-EXH005_icemod_']

for year in range(2002, 2015):
    L=365
    year_fake=2002
    if (year%4==0):
        L=366
    base1 = datetime.datetime(year_fake, 1, 5)
    date_list1 = [base1 + datetime.timedelta(days=x) for x in range(0, L, 5)]
    #
    base2 = datetime.datetime(year_fake, 1, 1)
    date_list2 = [base2 + datetime.timedelta(days=x) for x in range(0, L, 5)]
    for i in range(len(date_list1)):
        str1=date_list1[i].strftime('m%md%d.nc')
        str2=date_list2[i].strftime('m%md%d.nc')
        str1='y%s%s'%(year,str1)
        str2='y%s%s'%(year,str2)
        for j in range(5):
            print('mv'+' '+names[j]+str1+' '+names[j]+str2)
```

# Fix 2004

```python
names = ['ANHA4-EXH001_gridT_', 'ANHA4-EXH001_gridU_', 'ANHA4-EXH001_gridV_',
'ANHA4-EXH001_gridW_', 'ANHA4-EXH001_icemod_']

L = 366 + 365 + 365 + 365 - 5
base1 = datetime.datetime(2004, 3, 2)
date_list1 = [base1 + datetime.timedelta(days=x) for x in range(0, L, 5)]

base2 = datetime.datetime(2004, 3, 1)
date_list2 = [base2 + datetime.timedelta(days=x) for x in range(0, L, 5)]
for i in range(len(date_list1)):
    str1=date_list1[i].strftime('y%Ym%md%d.nc')
    str2=date_list2[i].strftime('y%Ym%md%d.nc')
    for j in range(5):
        print('mv'+' '+names[j]+str1+' '+names[j]+str2)
```
